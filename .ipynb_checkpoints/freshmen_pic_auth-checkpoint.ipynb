```python
import base64
import time
from io import BytesIO
import random
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from PIL import Image
from selenium.webdriver import ActionChains
from selenium.webdriver.chrome.options import Options

def show_pic(driver):
    pic_now = driver.get_screenshot_as_base64()
    Image.open(BytesIO(base64.b64decode(pic_now))).show()

chrome_options = Options()
chrome_options.add_argument('--no-sandbox') #解决DevToolsActivePort文件不存在的报错
chrome_options.add_argument('window-size=1920x3000') #指定浏览器分辨率
chrome_options.add_argument('--hide-scrollbars') #隐藏滚动条, 应对一些特殊页面
chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu') #谷歌文档提到需要加上这个属性来规避bug
driver = webdriver.Chrome(chrome_options=chrome_options)
driver.set_window_size(1200, 800)
```

```python
url = 'https://xmuxg.xmu.edu.cn/login'
driver.get(url)
time.sleep(1)
```

```python
show_pic(driver)
```

```python

imgs = []
for i in range(9):
    imgs.append(np.load(f"./pic_arr/{i}.npy"))

while True:
    try:
        time.sleep(2)
        bg_img = Image.open(
            BytesIO(base64.b64decode(driver.find_element(By.ID, "bg-img").get_attribute("src").split(",")[1])))
        sl_img = Image.open(
            BytesIO(base64.b64decode(driver.find_element(By.ID, "slider-img").get_attribute("src").split(",")[1])))
        dragger = driver.find_element(By.ID, 'slider-move-btn')
        bg_arr = np.asarray(bg_img)
        sl_arr = np.asarray(sl_img)
        score = 0
        num = -1
        for i in range(len(imgs)):
            n = np.sum(imgs[i] == bg_arr)
            if n > score:
                score = n
                num = i
        print("No." + str(num) + " pic")

        tar = imgs[num]
        x_l = 10000
        x_r = -1
        y_d = 10000
        y_u = -1
        for row in range(len(bg_arr)):
            for col in range(len(bg_arr[0])):
                if not (bg_arr[row][col][0] == tar[row][col][0] and bg_arr[row][col][1] == tar[row][col][1] and
                        bg_arr[row][col][2] == tar[row][col][2]):
                    if col < x_l: x_l = col
                    if col > x_r: x_r = col
                    if row < y_d: y_d = row
                    if row > y_u: y_u = row
        x_bg = (x_l + x_r) / 2
        y_bg = (y_u + y_d) / 2
        print(x_bg, y_bg)
        # the key is the x_bg not the y_bg

        # default the x of the slider is the centre
        x_sl = 55.5

        # the distance need to be dragged (in pixel)
        distance = x_bg - x_sl

        # the main axis is from 55 to 535(590-55)
        # the range of the pixel in win11 chrome is 206
        # so 1 pixel of distance means 535/206 = 0.385047 pixel in the screen
        distance_for_drag = distance * 0.385047
    except:
        break
    ActionChains(driver).click_and_hold(dragger).perform()

    # distance_for_drag += 15
    distance_for_drag *= 1.15
    pos = 0
    while True:
        rand_y = random.randint(-10, 10)
        if distance_for_drag - pos < 21:
            ActionChains(driver).move_by_offset(distance_for_drag - pos, rand_y).perform()
            break
        else:
            rand_x = random.randint(10, 20)
            print(pos, rand_x)
            pos += rand_x
            ActionChains(driver).move_by_offset(rand_x, rand_y).perform()
        time.sleep(random.random() / 10)

    show_pic(driver)
    ActionChains(driver).release(dragger).perform()
    time.sleep(0.5)
```

```python
import re
from datetime import datetime
time.sleep(1)
choose = []
last_time = None
spans = driver.find_elements(By.TAG_NAME, "span")
for span in spans:
    if span.get_attribute("title") == "请选择":
        choose.append(span)
    if re.match("....-..-.. ..:..:..", span.get_attribute("title"))!=None:
        last_time = span.get_attribute("title")
        print("last time: "+last_time)
last_time = datetime.strptime(last_time, "%Y-%m-%d %H:%M:%S")
now_time = datetime.now()
if last_time.year == now_time.year and last_time.month == now_time.month and last_time.day == now_time.day:
    print("已经打卡")
# 如果今天还没有打卡过 那么就再对于今天进行打卡
else:
    print("还没打卡")
    choose[-3].click()

    # choose yes
    time.sleep(1)
    labels = driver.find_elements(By.TAG_NAME, "label")
    yes_tag = None
    for label in labels:
        # print(label.get_attribute("title"))
        if label.get_attribute("title") == "是 Yes":
            yes_tag = label
    yes_tag.click()

    # 点击保存按键
    save = None
    for span in driver.find_elements(By.TAG_NAME, "span"):
        if span.get_attribute("class") == "form-save position-absolute":
            save = span
            break
    save.click()
    # show_pic(driver)
    # accept the alert
    driver.switch_to.alert.accept()
    show_pic(driver)
driver.close()
```
